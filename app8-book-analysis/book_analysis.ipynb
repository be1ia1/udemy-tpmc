# How many chapters?

```python
book.count('Chapter')
```

### With regex

```python
import re
```

```python
pattern = re.compile('Chapter \d+')
findings = re.findall(pattern, book)
len(findings)
```

# Which are the sentences where 'love' was used?

```python
pattern = re.compile('[A-Z]{1}[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^.]*.')
findings = re.findall(pattern, book)
findings
```

# What are the most used words?

### With collection

```python
pattern = re.compile('[a-zA-Z]+')
findings = re.findall(pattern, book.lower())
from collections import Counter
counterobj = Counter(findings)
counterobj.most_common()
```

### With dict

```python
d = {}
for word in findings:
    if word in d.keys():
        d[word] = d[word] + 1
    else:
        d[word] = 1
sorted(d.items(), key=lambda item: item[1], reverse=True)[0][0]
```

```python
def find_word(search):
    pattern = re.compile('[a-zA-Z]+')
    findings = re.findall(pattern,book.lower())

    answer_dict = {}

    for word in findings:
        if word in answer_dict.keys():
            answer_dict[word] += 1
        else:
            answer_dict[word] = 1
    
    return answer_dict.get(search, f'The book does not contain the word "{search}"')

print(find_word('love'))
print(find_word('hate'))
```
